```python
#QUESTION 9
def catPrice(price):
    if price <= 301:
        return '0-300'
    elif (price >= 300) & (price <= 500):
        return '301 - 500'
    elif price >= 501:
        return '500++'

df['price'] = df['price'].map(catPrice)
```
